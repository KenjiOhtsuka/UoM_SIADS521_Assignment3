## Visualization Technique

### Description of each visualization type

### How the visualizations complement each other and when each should be used.

In the demonstration we use **line charts**, **scatter plots**, **violin plots** including bar plots and jittered points, **bar charts** and **histograms**. Bar charts are used only in Explorative Data Analysis in advance of the demonstration. Each visualization serves a different purpose, and together they provide a complete view of long‑term trends, regional differences, and relationships among variables.

**Line charts** are used to show how indicators change over time.
They are effective for identifying long‑term trends, comparing continents, and observing periods of rapid improvement or decline.

**Scatter plots** reveal relationships between variables with actual data points.
Scatter plots also use color and point size to represent continent and population, helping users compare categories at a glance.

**Violin plots**, combined with jittered points and box plots, summarize the distribution of a variable within each category.
Jittered points show individual points, box plots highlight the median and interquartile range, and the violin shape shows the overall density. This combination provides a detailed view of variation within each category.

**Bar charts** are helpful when the goal is to compare magnitudes rather than trends or relationships.

**Histograms** show the distribution of a variable. They help users understand how values are spread globally and whether the distribution is narrow, wide, or skewed.

### How the visualizations are combined into a dashboard and how they complement each other.

The dashboard contains Line charts, Histograms, and Scatter plots.
Each visualization highlights a different aspect of the Gapminder dataset, and together they provide a clear and interactive view of global health and economic development.

**Line charts** show how life expectancy changes over time for the selected country.
A vertical red line marks the chosen year, helping users connect long‑term trends with the specific year shown in the other charts.

**Histograms** display the global distribution of life expectancy in the selected year.
The bin corresponding to the selected country is highlighted in red, making it easy to see where that country falls within the worldwide distribution.

**Scatter plots** show the relationship between GDP per capita and life expectancy for the selected year.
The chosen country is emphasized with a red star marker, allowing users to compare it directly with other countries.

These visualizations complement each other:

- **Line charts** show **trends over time**,
- **Histograms** show **global distributions**,
- **Scatter plots** show **relationships**.

With ipywidgets, all charts update simultaneously when the user changes the country or year.
This coordinated interactivity allows users to explore long‑term trends, compare countries, and understand global patterns within a single dashboard.

### Plotly

Plotly is an open-source graphing library developed by Plotly Technologies Inc., with contributions from the open-source community. It allows for the creation of interactive and visually appealing plots. It supports a wide range of chart types, including scatter plots, line charts, bar charts, and more. Additionally, Plotly integrates well with Jupyter notebooks, making it easy to create and display visualizations within the notebook environment.

Plotly follows a hybrid declarative and imperative design. Users can specify visual attributes declaratively through figure objects while fine-grained layout adjustments and callbacks are handled imperatively. This makes Plotly flexible for both simple and advanced visualizations.

The limitations include a steeper learning curve for detailed customization and potential performance issues with very large datasets. However, its interactivity and customization options make it a powerful tool for data visualization.

It is suitable for this assignment because it provides rich interactivity, it can be used in Jupyter Notebooks, it can create multiple types of visualizations.

### Life expectancy over time

According to the correlation analysis, life expectency and year are moderately and positively correlated (0.43).
We can see the trend in line chart below.

```python
import plotly.express as px
import plotly.graph_objects as go

base_colors = px.colors.qualitative.Plotly
continents = df["continent"].unique()
continent_colors = {
    cont: base_colors[i % len(base_colors)]
    for i, cont in enumerate(continents)
}

def hex_to_rgba(hex_color, alpha=0.3):
    hex_color = hex_color.lstrip('#')
    r = int(hex_color[0:2], 16)
    g = int(hex_color[2:4], 16)
    b = int(hex_color[4:6], 16)
    return f"rgba({r},{g},{b},{alpha})"
```

```python

fig = go.Figure()

for country, d in df.groupby("country"):
    cont = d["continent"].iloc[0]
    base_color = continent_colors[cont]
    rgba_color = hex_to_rgba(base_color, alpha=0.3)

    fig.add_trace(go.Scatter(
        x=d["year"],
        y=d["lifeExp"],
        mode="lines",
        line=dict(color=rgba_color, width=2),
        name=country,
        legendgroup=cont,
        showlegend=False
    ))

# continent ごとの凡例を追加
for cont, base_color in continent_colors.items():
    fig.add_trace(go.Scatter(
        x=[None], y=[None],
        mode="lines",
        line=dict(color=base_color, width=3),
        name=cont
    ))

fig.update_layout(
    title="Life Expectancy Over Time",
    xaxis_title="Year",
    yaxis_title="Life Expectancy"
)

fig.show()
```

It shows the life expectency has been increasing over time for most of the countries, but we can see decreasing trend for some countries and years. It seems that life expectency in Oceania and European countries is higher and one in African countries lower. 

Which countries have the highest life expectency increase between 1952 and 2007?

### 1. Introduction

### 2. Country-Level Trends: Improvements in Life Expectancy

Countries show different patterns in how life expectancy changes over time.
**Japan** increases steadily from the 1950s as healthcare improves.
**Zambia** and **Zimbabwe** show declines in the 1990s due to health crises, followed by slow recovery.
**Oman** shows one of the largest improvements, rising from very low levels to near high‑income countries.

These country-level differences motivate a broader look at regional patterns.

To illustrate how life expectancy has evolved at the country level,
I provide an interactive line chart that allows the user to select any country.
This makes it possible to explore differences in development pace and historical context.
For example, selecting Japan shows a rapid increase in life expectancy after the 1950s.

```python
import ipywidgets as widgets
from ipywidgets import interact
import plotly.express as px

country_options = {
    f"{row['continent']} - {row['country']}": row['country']
    for _, row in df[['continent','country']].drop_duplicates().iterrows()
}
country_options = dict(sorted(country_options.items(), key=lambda item: item[0]))

# Create a dropdown with Japan as the default value
country_selector = widgets.Dropdown(
    options=country_options,
    value="Japan",
    description='Country:',
    style={'description_width': 'initial'}
)

@interact(country=country_selector)
def plot_country_life_expectancy(country):
    df_country = df[df['country'] == country]
    fig = px.line(
        df_country,
        x="year",
        y="lifeExp",
        title=f"Life Expectancy Over Time: {country}",
        labels={"lifeExp": "Life Expectancy", "year": "Year"}
    )
    fig.update_layout(autosize=False)
    fig.update_yaxes(
        range=[
            df["lifeExp"].min() * 0.95,
            df["lifeExp"].max() * 1.05
        ]
    )
    fig.show()
```

These country‑level patterns motivate a broader question:
how do these trends differ across continents?  

### 3. Continental Patterns: Regional Differences in Development

The previous section highlights how individual countries have changed over time.
Examining broader regional patterns helps explain why these trends differ across the world.
Continents vary widely in economic development, demographic growth, and public‑health conditions,
and these structural differences shape life expectancy outcomes.

Aggregating the data by continent reveals clear regional differences.

- **Europe** and **Oceania** maintain high life expectancy throughout the period.

GDP per capita shows a similar pattern: regions with faster economic growth tend to improve life expectancy more quickly. Population trends also show that Asia and Africa drive most global population growth.

(Continent-level GDP per capita is calculated as the total GDP of a continent divided by its total population. Continent-level life expectancy is calculated as the simple average of life expectancies across countries within that continent. ### 4. Global Inequality: Are Countries Becoming More Similar Over Time?

The continental differences observed in Section 3 show that regions follow distinct development patterns.
These differences help explain part of the inequality seen within and across regions.

To assess global inequality directly, we examine the standard deviation of life expectancy across all countries over time.


```python
df_std_lifeExp = df.groupby("year")["lifeExp"].std().reset_index()

fig = px.line(
    df_std_lifeExp,
    x="year",
    y="lifeExp",
    title="Global Life Expectancy Inequality Over Time (Standard Deviation)",
    labels={"lifeExp": "Life Expectancy Standard Deviation", "year": "Year"}
)
fig.update_layout(width=900, height=500)
fig.show()
```

The global standard deviation does not follow a simple downward trend.
It fell from the 1950s to the mid‑1980s, rose again through the 1990s and early 2000s, and then became slightly more stable in recent years.
This means countries sometimes become more similar and sometimes more different.

The standard deviation alone does not tell us which regions contribute most to global inequality.
However, the trends in Section 3 suggest that slower improvements in parts of Africa help keep these gaps in place.

GDP per capita shows a similar pattern of regional differences.
This suggests that economic conditions and health outcomes are related,
although their global inequality patterns do not always move in the same direction.

```python
df_std_gdp = df.groupby("year")["gdpPercap"].std().reset_index()

fig = px.line(
    df_std_gdp,
    x="year",
    y="gdpPercap",
    title="Global GDP per Capita Inequality Over Time (Standard Deviation)",
    labels={"gdpPercap": "GDP per Capita Standard Deviation", "year": "Year"}
)
fig.update_layout(width=900, height=500)
fig.show()
```

This indicates that economic inequality, like health inequality, has experienced periods of convergence and divergence rather than a steady decline.

Together, these findings indicate that although global health has improved overall,
but the pace of improvement still varies widely across regions.

### 5. Relationships Between Indicators: What Drives Life Expectancy?

To understand the drivers behind the inequalities shown in Section 4,
this section analyzes the relationships among GDP per capita, life expectancy, and population.

Scatter plots show the relationships between key variables:

- **GDP per capita vs. Life Expectancy** shows a strong positive association.
  Wealthier countries tend to have longer life expectancy due to better healthcare systems, infrastructure, and living conditions.
- **Population vs. GDP per capita** shows almost no meaningful relationship.
  Large populations do not imply higher or lower economic development.
- **Population vs. Life Expectancy** also shows weak correlation.
  Population size is influenced by fertility rates, age structure, and migration rather than health outcomes.


```python
fig = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    color="continent",
    hover_name="country",
    title=f"GDP per Capita vs Life Expectancy (correlation: {df['gdpPercap'].corr(df['lifeExp']):.2f})",
    labels={"gdpPercap": "GDP per Capita", "lifeExp": "Life Expectancy"},
    opacity=0.5,
    log_x=True
)
fig.update_layout(width=900, height=500)
fig.show()
```

```python
fig = px.scatter(
    df,
    x="pop",
    y="gdpPercap",
    color="continent",
    hover_name="country",
    title=f"Population vs GDP per Capita (correlation: {df['pop'].corr(df['gdpPercap']):.2f})",
    labels={"pop": "Population", "gdpPercap": "GDP per Capita"},
    opacity=0.5,
    log_x=True,
    log_y=True
)
fig.update_layout(width=900, height=500)
fig.show()
```

```python
fig = px.scatter(
    df,
    x="pop",
    y="lifeExp",
    color="continent",
    hover_name="country",
    title=f"Population vs Life Expectancy (correlation: {df['pop'].corr(df['lifeExp']):.2f})",
    labels={"pop": "Population", "lifeExp": "Life Expectancy"},
    opacity=0.5,
    log_x=True
)
fig.update_layout(width=900, height=500)
fig.show()
```

These results confirm that economic development, not population size, is the primary driver of improvements in life expectancy.
